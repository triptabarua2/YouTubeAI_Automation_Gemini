## ধাপ ৩: API Keys লোড করুন (Colab Secrets থেকে)
আপনার API Key গুলো নিরাপদে রাখার জন্য Google Colab-এর **Secrets** ফিচারটি ব্যবহার করুন।

**কিভাবে Secrets সেট করবেন:**
১. কোলাবের বাম পাশের মেনু থেকে **চাবি (Key)** আইকনটিতে ক্লিক করুন (এটি 'Secrets' অপশন)।
২. **'Add new secret'** বাটনে ক্লিক করুন।
৩. নিচের নামগুলো ব্যবহার করে আপনার API Key গুলো যোগ করুন এবং প্রতিটি কী-এর পাশের **'Notebook access'** বাটনটি অন (নীল) করে দিন:
- `GOOGLE_API_KEY`
- `ELEVENLABS_API_KEY`
- `TELEGRAM_BOT_TOKEN`
- `TELEGRAM_CHAT_ID`

নিচের সেলটি রান করলে এই কী-গুলো আপনার প্রোজেক্টে লোড হয়ে যাবে।

```python
from google.colab import userdata
import os

try:
    # Colab Secrets থেকে API Key গুলো লোড করা
    os.environ["GOOGLE_API_KEY"] = userdata.get("GOOGLE_API_KEY")
    os.environ["ELEVENLABS_API_KEY"] = userdata.get("ELEVENLABS_API_KEY")
    os.environ["TELEGRAM_BOT_TOKEN"] = userdata.get("TELEGRAM_BOT_TOKEN")
    os.environ["TELEGRAM_CHAT_ID"] = str(userdata.get("TELEGRAM_CHAT_ID"))
    print("✅ API Keys successfully loaded from Colab Secrets!")
except Exception as e:
    print(f"⚠️ Error loading secrets: {e}")
    print("নিশ্চিত করুন যে আপনি Colab Secrets-এ সঠিক নাম ব্যবহার করেছেন এবং 'Notebook access' অন করেছেন।")
```
