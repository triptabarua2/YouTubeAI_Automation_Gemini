## ধাপ ৩: API Keys সেটআপ (Colab Secrets)
আপনার API Key গুলো নিরাপদে রাখার জন্য Google Colab-এর **Secrets** ফিচারটি ব্যবহার করুন।

**কিভাবে Secrets সেট করবেন:**
১. কোলাবের বাম পাশের মেনু থেকে **চাবি (Key)** আইকনটিতে ক্লিক করুন (এটি 'Secrets' অপশন)।
২. **'Add new secret'** বাটনে ক্লিক করুন।
৩. নিচের নামগুলো ব্যবহার করে আপনার API Key গুলো যোগ করুন এবং প্রতিটি কী-এর পাশের **'Notebook access'** বাটনটি অন (নীল) করে দিন:
- `GROQ_API_KEY` (প্রয়োজনীয়)
- `GOOGLE_API_KEY` (প্রয়োজনীয়)
- `ELEVENLABS_API_KEY` (প্রয়োজনীয়)
- `TELEGRAM_BOT_TOKEN` (ঐচ্ছিক)
- `TELEGRAM_CHAT_ID` (ঐচ্ছিক)

**নোট:** এখন থেকে `config.py` ফাইলটি অটোমেটিক এই সিক্রেটসগুলো লোড করে নিবে, তাই আলাদা করে কোড লেখার প্রয়োজন নেই।

## ধাপ ৪: YouTube OAuth JSON ফাইলগুলো আপলোড করুন
আপনার `client_secret_channel1.json`, `client_secret_channel2.json`, এবং `client_secret_channel3.json` ফাইলগুলো আপলোড করতে হবে।

১. বাম পাশের **Files (ফোল্ডার আইকন)** এ ক্লিক করুন।
২. `YouTubeAI_Automation_Gemini` ফোল্ডারের ভেতরে সরাসরি ফাইলগুলো ড্র্যাগ করে বা আপলোড বাটনে ক্লিক করে দিয়ে দিন।

**গুরুত্বপূর্ণ:** কোলাব সেশন বন্ধ হলে এই ফাইলগুলো মুছে যেতে পারে। তাই সবচেয়ে ভালো হয় যদি আপনি ফাইলগুলো আপনার Google Drive-এ একটি ফোল্ডারে রেখে সেটি কোলাবের সাথে মাউন্ট (Mount) করে দেন।
